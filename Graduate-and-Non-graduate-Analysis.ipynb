```python
# This line of code calls the 'plot_graduate_premium' method on the 'obj_int' instance.
# The 'plot_graduate_premium' method is designed to create and display a bar plot visualizing
# the average graduate premium by major category. The method takes 'sorted_premiums' as an argument,
# which contains the sorted average graduate premiums for each major category.
# This plot is useful for visual analysis, helping to compare and understand the economic
# value of higher education across different fields of study.
obj_int.plot_graduate_premium(sorted_premiums)

# After executing this line, a bar plot will be displayed showing the sorted average graduate premiums.
# This visualization aids in identifying which major categories offer higher or lower financial benefits to graduates.
```
